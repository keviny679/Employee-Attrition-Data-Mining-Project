```python
import pandas as pd
import numpy as np

df = pd.read_csv('IBM_Employee_Attrition.csv')

def preprocess(df):
    
    # 1. Map binary categorical columns (Yes/No, Male/Female)
    def map_binary_column(column, map_dict):
        df[column] = df[column].map(map_dict)
    
    # 3. Map ordinal categorical columns
    def map_ordinal_column(column, map_dict):
        df[column] = df[column].map(map_dict)
    
    map_binary_column('Attrition', {'Yes': 1, 'No': 0})
    map_binary_column('Gender', {'Male': 0, 'Female': 1})
    map_binary_column('OverTime', {'Yes': 1, 'No': 0})
    
    # Mapping Businesstravel 
    map_ordinal_column('BusinessTravel', {
        'Non-Travel': 0, 
        'Travel_Rarely': 1, 
        'Travel_Frequently': 2
    })

    # Mapping Department
    map_ordinal_column('Department', {
        'Research & Development': 0,
        'Sales': 1,
        'Human Resources': 2
    })

    # Mapping Educationfield
    map_ordinal_column('EducationField', {
        'Life Sciences': 0,
        'Medical': 1,
        'Technical Degree': 2,
        'Marketing': 3,
        'Other': 4
    })

    # Mapping all Jobs
    map_ordinal_column('JobRole', {
        'Sales Executive': 0,
        'Research Scientist': 1,
        'Laboratory Technician': 2,
        'Manufacturing Director': 3,
        'Healthcare Representative': 4,
        'Manager': 5,
        'Human Resources': 6,
        'Research Director': 7
    })

    # Mapping Marriage status
    map_ordinal_column('Department', {
        'Married': 0,
        'Single': 1,
        'Divorced': 2
    })
    
    # Dropping the rates since monthly income is better data
    df = df.drop(columns='DailyRate') 
    df = df.drop(columns='HourlyRate')
    df = df.drop(columns='MonthlyRate')

    df = df.drop(columns='EmployeeCount') # Dropped since whole column is 1

    df = df.drop(columns='Over18') # this column is useless, all over 18
    
    return df

preprocess(df)
```
