```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IBM_Employee_Attrition.csv')
```

```python
def preprocess(df):
    
    # Dropping the rates since monthly income is better data
    df = df.drop(columns='DailyRate') 
    df = df.drop(columns='HourlyRate')
    df = df.drop(columns='MonthlyRate')

    df = df.drop(columns='BusinessTravel') # Similar to job involvement

    df = df.drop(columns='EmployeeCount') # Dropped since whole column is 1

    df = df.drop(columns='Over18') # this column is useless, all over 18
    
    df = df.drop(columns='Education') # Dropping education since the education field is more important

    df = df.drop(columns='StandardHours') # Dropping since whole column is 80 so its irrelevant

    df = df.drop(columns='StockOptionLevel') # Dropping since I feel its not too related

    df = df.drop(columns='TrainingTimesLastYear') # Dropping because if they got the job the training went fine
    
    return df

preprocess(df)
```
